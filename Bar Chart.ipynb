# Plotting Bar Chart


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class_student1 = [30, 10, 25, 20, 10]
class_student2 = [40, 5, 20, 20, 10]
class_student3 = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes, class_student1)
```

```python
plt.barh(classes, class_student1)
```

```python
plt.bar(classes, class_student1, width=0.5, align="edge", color="red")
```

```python
plt.bar(classes, class_student1, width=0.5, align="edge", color="red", edgecolor='k', linewidth=2)
```

```python
plt.bar(classes, class_student1, width=0.5, align="edge", color="red", edgecolor='k', linewidth=2, linestyle=":")
```

```python
plt.bar(classes, class_student1, width=0.5, align="edge", color="red", edgecolor='k', linewidth=2, linestyle=":",
        label="Class Student 1",visible=False)
plt.legend()
```

```python
plt.bar(classes, class_student1, width=0.5, align="edge", color="red", edgecolor='k', linewidth=2, linestyle=":",
        label="Class Student 1",visible=True)
plt.legend()
```

```python
style.use("ggplot")
plt.figure(figsize=(18,9))

plt.bar(classes, class_student1, width=0.5, align="edge", color="red", edgecolor='k', linewidth=2, linestyle=":",
        label="Class Student 1",visible=True)
plt.legend()
```

```python

plt.barh(classes, class_student1, align="edge", color="red", edgecolor='k', linewidth=2, linestyle=":",
        label="Class Student 1",visible=True)
plt.legend()
```

```python

plt.bar(classes, class_student1, width=0.5, align="edge", color="red", edgecolor='k', linewidth=2, linestyle=":",
        label="Class Student 1",visible=True)
plt.title("Bar Chart of XYZ Class")
plt.xlabel("Classes")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```

```python

plt.bar(classes, class_student1, width=0.5, align="edge", color="red", edgecolor='k', linewidth=2, linestyle=":",
        label="Class Student 1",visible=True)
plt.title("Bar Chart of XYZ Class")
plt.xlabel("Classes", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
plt.show()
```

```python
plt.bar(classes, class_student1, width=0.2, color="b", edgecolor='k', label="Class Student 1")

plt.bar(classes, class_student2, width=0.2, color="g", edgecolor='k', label="Class Student 2")

plt.title("Bar Chart of XYZ Class")
plt.xlabel("Classes", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
plt.show()
```

```python
classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, class_student1, width, color="b", edgecolor='k', label="Class Student 1")

plt.bar(classes_index + width, class_student2, width, color="g", edgecolor='k', label="Class Student 2")

plt.xticks(classes_index + width, classes)
plt.title("Bar Chart of XYZ Class")
plt.xlabel("Classes", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
plt.show()
```

```python
classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, class_student1, width, color="b", edgecolor='k', label="Class Student 1")

plt.bar(classes_index + width, class_student2, width, color="g", edgecolor='k', label="Class Student 2")

plt.xticks(classes_index + width, classes, rotation = 30)
plt.title("Bar Chart of XYZ Class")
plt.xlabel("Classes", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
plt.show()
```

```python
classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, class_student1, width, color="b", edgecolor='k', label="Class Student 1")

plt.bar(classes_index + width, class_student2, width, color="g", edgecolor='k', label="Class Student 2")

plt.bar(classes_index + width, class_student3, width, color="y", edgecolor='k', label="Class Student 2")

plt.xticks(classes_index + width, classes, rotation = 30)
plt.title("Bar Chart of XYZ Class")
plt.xlabel("Classes", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
plt.show()
```
