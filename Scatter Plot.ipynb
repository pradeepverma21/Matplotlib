# Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("googleplaystore[1].csv",nrows = 1000)
```

```python
df.shape
```

```python
df.head()
```

```python
x = df["Rating"]
y = df["Reviews"]
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
```

```python

plt.scatter(x, y, c='r')
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x, y, c='r')
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
```

```python

plt.scatter(x, y, c='r', marker='+')
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
```

```python

plt.scatter(x, y, c='r', marker='+', s=100)
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
```

```python

plt.scatter(x, y, c='r', marker='+', s=100, alpha=0.5, linewidth=5)
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
```

```python

plt.scatter(x, y, c='r', marker='*', s=100, alpha=0.5, linewidth=5, edgecolor='g')
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x, y, c='r', marker='*', s=100, alpha=0.5, linewidth=5, edgecolor='g')
plt.scatter(x, df["Installs"], c='y', marker='o', s=100, alpha=0.5, linewidth=5, edgecolor='c')
plt.title("Google Play Store App Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review & Installs")
```
