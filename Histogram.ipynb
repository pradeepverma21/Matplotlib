# Plotting Histogram 

```python
import matplotlib.pyplot as plt
import numpy as np
import random


ml_student_age = np.random.randint(18, 45, (100))
py_student_age = np.random.randint(15, 45, (100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("ML Student Age Histogram")
plt.xlabel("Student age categories")
plt.ylabel("Student Age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
```

```python
plt.hist(ml_student_age, bins, rwidth=0.8, orientation="horizontal", color='m', label="ML Student")
plt.title("ML Student Age Histogram")
plt.xlabel("Student age categories")
plt.ylabel("Student Age")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, orientation="vertical", color=['m', 'y'], label=["ML Student", "PY Student"])

# plt.hist(py_student_age, bins, rwidth=0.8, orientation="vertical", color='y', label="PY Student")

plt.title("ML Student Age Histogram")
plt.xlabel("Student age categories")
plt.ylabel("Student Age")
plt.legend()
plt.show()
```
