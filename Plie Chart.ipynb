# Pie Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
classes = ["Python", "R", "Artificial Intelligence", "Machine Learnig", "Data Sciece"]
classes_student = [45, 15, 35, 25, 30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(classes_student)
plt.show()
```

```python
plt.pie(classes_student, labels=classes)
plt.show()
```

```python
explode = [0.2, 0, 0, 0.2, 0]
plt.pie(classes_student, labels=classes, explode=explode)
plt.show()
```

```python
explode = [0.2, 0, 0, 0.2, 0]
color = ['c', 'b', 'r', 'k','g']
plt.pie(classes_student, labels=classes, explode=explode, colors = color)
plt.show()
```

```python
explode = [0.2, 0, 0, 0.2, 0]
color = ['c', 'b', 'r', 'k','g']
plt.pie(classes_student, labels=classes, explode=explode, colors = color, autopct = "%0.1f%%")
plt.show()
```

```python
explode = [0.2, 0, 0, 0.2, 0]
color = ['c', 'b', 'r', 'k','g']
plt.pie(classes_student, labels=classes, explode=explode, colors = color, autopct = "%0.1f%%", shadow=True)
plt.show()
```

```python
explode = [0.2, 0, 0, 0.2, 0]
color = ['c', 'b', 'r', 'k','g']
plt.pie(classes_student, labels=classes, explode=explode, colors = color, autopct = "%0.1f%%", shadow=True, radius=1.2)
plt.show()
```

```python
explode = [0.2, 0, 0, 0.2, 0]
color = ['c', 'b', 'r', 'k','g']
text = {"fontsize":15}
plt.pie(classes_student, labels=classes, explode=explode, colors = color, autopct = "%0.1f%%", shadow=True, radius=1.2, 
       textprops=text)
plt.show()
```

```python
plt.pie(classes_student, labels=classes, explode=explode, colors = color, autopct = "%0.1f%%", shadow=True, radius=1.2, 
       textprops=text, startangle=90)
plt.show()
```

```python
plt.pie(classes_student, labels=classes, explode=explode, colors = color, autopct = "%0.1f%%", shadow=True, radius=1.2, 
       textprops=text, startangle=90)
plt.legend(loc=(-1,0))
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,4))

colors = ['r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w']
labels = np.ones(20)

plt.pie([1], colors='k', radius=2.05)
plt.pie(labels, colors=colors, radius=2.0)

plt.pie([1], colors='g', radius=1.8)
plt.pie([1], colors='y', radius=1.6)
plt.pie([1], colors='c', radius=1.3)
plt.pie([1], colors='b', radius=1.1)
plt.pie([1], colors='m', radius=0.9)

plt.pie([1], colors='b', radius=0.32)
plt.pie(labels, colors=colors, radius=0.3)

plt.pie([1], colors='w', radius=0.2)
plt.pie([1], colors='k', radius=0.1)

plt.show()
```
